```python
import os
from pathlib import Path
from utils import data_loader_utils
from utils import user_defined_functions as udf
import itertools 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import pandas as pd
import math
from scipy import ndimage
from scipy.stats import pearsonr
import scipy.signal as signal
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
file1 = Path("./data/M01/OP07/good/M01_Feb_2019_OP07_000.h5").absolute()
file4 = Path("./data/M01/OP07/good/M01_Aug_2021_OP07_000.h5").absolute()
file2 = Path("./data/M01/OP07/good/M01_Aug_2019_OP07_000.h5").absolute()
file3 = Path("./data/M01/OP07/good/M01_Feb_2021_OP07_000.h5").absolute()
file5 = Path("./data/M01/OP07/bad/M01_Feb_2019_OP07_000.h5").absolute()

x = data_loader_utils.datafile_read(file1)
y = data_loader_utils.datafile_read(file2)
z = data_loader_utils.datafile_read(file3)
w = data_loader_utils.datafile_read(file4)
bad = data_loader_utils.datafile_read(file5)
```

```python
Fs = 2000
dt = 1/Fs

def max_psd_in_band(signal, Fs, fmin=180, fmax=220):
    Pxx, freqs = plt.psd(signal, NFFT=len(signal), Fs=Fs)
    plt.close()  # prevent plotting

    band_mask = (freqs >= fmin) & (freqs <= fmax)
    return np.max(Pxx[band_mask]), freqs[band_mask][np.argmax(Pxx[band_mask])]


plt.figure(figsize=(20, 5))
plt.psd(x[:,0], NFFT = len(x), Fs = Fs)
plt.show()

max_val, freq_at_max = max_psd_in_band(x[:,0], Fs)
print(f": max PSD = {10*np.log10(max_val):.4e} at {freq_at_max:.2f} Hz")


plt.figure(figsize=(20, 5))
plt.psd(y[:,0], NFFT = len(y), Fs = Fs)
plt.show()

max_val, freq_at_max = max_psd_in_band(y[:,0], Fs)
print(f": max PSD = {10*np.log10(max_val):.4e} at {freq_at_max:.2f} Hz")

plt.figure(figsize=(20, 5))
plt.psd(z[:,0], NFFT = len(z), Fs = Fs)
plt.show()

max_val, freq_at_max = max_psd_in_band(z[:,0], Fs)
print(f": max PSD = {10*np.log10(max_val):.4e} at {freq_at_max:.2f} Hz")

plt.figure(figsize=(20, 5))
plt.psd(w[:,0], NFFT = len(w), Fs = Fs)
plt.show()

max_val, freq_at_max = max_psd_in_band(w[:,0], Fs)
print(f": max PSD = {10*np.log10(max_val):.4e} at {freq_at_max:.2f} Hz")

```

```python
import numpy as np
from scipy import signal



# --- Application and Tuning ---
# Load your signal (e.g., 132 seconds of OP00)
# original_data = load_data(...) 

# TUNE FREQUENCY HERE
ORIGINAL_FS = 2000  # Fixed based on the report 
TARGET_FS = 1000    # Change this to 500, 800, etc.

downsampled_signal_x_x_axis = udf.resample_milling_data(x[:,0], ORIGINAL_FS, TARGET_FS)
downsampled_signal_x_y_axis = udf.resample_milling_data(x[:,1], ORIGINAL_FS, TARGET_FS)
downsampled_signal_y_x_axis = udf.resample_milling_data(y[:,0], ORIGINAL_FS, TARGET_FS)
downsampled_signal_y_y_axis = udf.resample_milling_data(y[:,1], ORIGINAL_FS, TARGET_FS)
downsampled_signal_z_x_axis = udf.resample_milling_data(z[:,0], ORIGINAL_FS, TARGET_FS)
downsampled_signal_z_y_axis = udf.resample_milling_data(z[:,1], ORIGINAL_FS, TARGET_FS)
downsampled_signal_w_x_axis = udf.resample_milling_data(w[:,0], ORIGINAL_FS, TARGET_FS)
downsampled_signal_w_y_axis = udf.resample_milling_data(w[:,1], ORIGINAL_FS, TARGET_FS)
downsampled_signal_bad_x_axis = udf.resample_milling_data(bad[:,0], ORIGINAL_FS, TARGET_FS)
downsampled_signal_bad_y_axis = udf.resample_milling_data(bad[:,1], ORIGINAL_FS, TARGET_FS)

# Look at PSD with downsampled measurements and see the difference between a good and a bad measurement, taken over the interval of the section
# where one movement in and out of the material is performed (this is normally, with a sample frequency of 1000 Hz, for process 7, between sample 2400
# and sample 6000)

plt.figure(figsize=(20, 5))
plt.psd(downsampled_signal_x_x_axis[2400:6000], NFFT = len(x), Fs = TARGET_FS)
plt.show()

max_val, freq_at_max = max_psd_in_band(downsampled_signal_x_x_axis[2400:6000], Fs = TARGET_FS)
print(f": max PSD = {10*np.log10(max_val):.4e} at {freq_at_max:.2f} Hz")

plt.figure(figsize=(20, 5))
plt.psd(downsampled_signal_bad_x_axis[2400:6000], NFFT = len(x), Fs = TARGET_FS)
plt.show()

max_val, freq_at_max = max_psd_in_band(downsampled_signal_bad_x_axis[2400:6000], Fs = TARGET_FS)
print(f": max PSD = {10*np.log10(max_val):.4e} at {freq_at_max:.2f} Hz")
```

```python
#Try with ARMAX model, but, as visible, no improvement

signal_used_armax = downsampled_signal_x_x_axis[2400:6500]

threshold = 0.001 * np.std(signal_used_armax)
u = (np.abs(signal_used_armax) > threshold).astype(float)
print(u)
# Step 2: Fit ARMAX model
p, q = 2, 3
model = SARIMAX(
    signal_used_armax,
    exog=u,
    order=(p, 0, q),
    enforce_stationarity=False,
    enforce_invertibility=False
)
results = model.fit(disp=False)

print(results.summary())

# Step 3: Analyze residuals
residuals = results.resid

plt.figure()
plt.plot(residuals)
plt.title("ARMAX Residuals")
plt.show()

plot_acf(residuals, lags=40)
plt.show()
```
