```python
import os
from pathlib import Path
from utils import data_loader_utils
import itertools 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import pandas as pd
import math
from scipy import ndimage
from statsmodels.tsa.stattools import adfuller
```

```python
file1 = Path("./data/M01/OP07/good/M01_Feb_2019_OP07_000.h5").absolute()
file2 = Path("./data/M01/OP07/bad/M01_Feb_2019_OP07_000.h5").absolute()
x = data_loader_utils.datafile_read(file1)
y = data_loader_utils.datafile_read(file2)
```

```python
print(x.shape)
print(y.shape)
print(len(y))
print(np.mean(x[:,0]))
```

```python
delay_range = range(-2000, 2000, 1)
cross_corr = np.zeros((len(delay_range),1))
for i in range(0, len(delay_range), 1):
    x_i = x[0:min(len(x),len(y)), 0]
    y_i = ndimage.shift(y[0:min(len(x),len(y)), 0], delay_range[i])
    cross_corr[i] = np.corrcoef(x_i,y_i)[0,1]
```

```python
plt.figure(figsize=(20, 5))
plt.plot(delay_range, cross_corr[:])
plt.xlim(0, 50)
#plt.locator_params(axis='x', nbins=len(delay_range))
plt.grid()
plt.show()
```
